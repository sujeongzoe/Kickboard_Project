### 강남구

```python
import requests
import pandas as pd
import numpy as np


def doro_to_coords(df):
    data = df.copy()
    data['경도'] = 0
    data['위도'] = 0
    
    for i in range(len(data)):
        if pd.notna(data.loc[i, '주소']) and data.loc[i, '주소'] != '탈퇴':  # '도로명주소' → '주소' 변경
            try:
                address = data.loc[i, '주소']
                url = f'https://dapi.kakao.com/v2/local/search/address.json?query={address}'
                headers = {"Authorization": "KakaoAK " + "39aa05f286da07187ee24e01563aa0f2"}
                response = requests.get(url, headers=headers).json()
                
                if response.get('documents'):
                    temp = response['documents'][0].get('road_address')  # 도로명주소 정보 가져오기
                    if temp:
                        data.loc[i, ['경도', '위도']] = [temp['x'], temp['y']]
            
            except Exception as e:
                print(f"오류 발생: {e} (행 {i})")
    
    return data

# CSV 파일 읽기 (경로 수정)
df = pd.read_csv(r"C:\Users\user\Desktop\0221 미팅\서울시 강남구 유치원 일반현황.csv", encoding='cp949')

# 함수 적용
result_df = doro_to_coords(df)

# 결과 저장 (경로 수정)
result_df.to_csv(r"C:\Users\user\Desktop\0221 미팅\서울시 강남구 유치원 일반현황_위경도.csv", encoding='utf-8-sig', index=False)
```

## 3. 경로당 데이터 위경도 변환

```python
import requests
import pandas as pd
import numpy as np
import re
import json

# 🔹 주소 → 경도, 위도 변환 함수
def doro_to_coords(df):
    data = df.copy().reset_index(drop=True)  
    data['경도'] = 0
    data['위도'] = 0
    
    # 🔥 주소에서 마지막 괄호 제거 & "서울특별시 [자치구명] " 추가
    data['주소'] = data.apply(lambda row: f"서울특별시 {row['자치구']} " + re.sub(r'\s*\(.*?\)$', '', str(row['Unnamed: 3'])) if pd.notna(row['Unnamed: 3']) else row['Unnamed: 3'], axis=1)
    
    for i in range(len(data)):
        if pd.notna(data.iloc[i]['주소']):  
            try:
                address = data.iloc[i]['주소']
                url = f'https://dapi.kakao.com/v2/local/search/address.json?query={address}'
                headers = {"Authorization": "KakaoAK " + "39aa05f286da07187ee24e01563aa0f2"}
                response = requests.get(url, headers=headers)
                response_json = json.loads(response.text.encode('utf-8').decode('utf-8'))

                if response_json.get('documents'):
                    temp = response_json['documents'][0].get('road_address') or response_json['documents'][0].get('address')
                    if temp:
                        data.loc[i, ['경도', '위도']] = [temp['x'], temp['y']]
                   
            except Exception as e:
                print(f"❌ 오류 발생: {e} (행 {i})")
    
    return data

# 🔹 엑셀 파일 읽기
file_path = r"C:\Users\user\Desktop\0221 미팅\열린데이터V2(경로당현황-최신자료3596개소)_2401.xlsx"

# 🔹 강남구, 마포구, 송파구 시트만 추출
sheets_to_extract = ['강남구', '마포구', '송파구']
dfs = pd.read_excel(file_path, sheet_name=sheets_to_extract)

# 🔹 각 시트별 데이터 처리 및 저장
for region, df in dfs.items():
    df = df.copy()
    
    # 🔹 컬럼명 표준화
    df.columns = df.columns.str.strip()
    
    # 🔹 '자치구' 칼럼 추가 (시트명 기준)
    df['자치구'] = region
    
    # 🔹 주소 → 경도, 위도 변환
    df_result = doro_to_coords(df)
    
    # 🔹 변환된 데이터 저장
    output_path = fr"C:\Users\user\Desktop\0221 미팅\{region}_경로당.csv"
    df_result.to_csv(output_path, encoding='utf-8-sig', index=False)
    
    # 🔹 변환 실패한 데이터 개수 확인
    zero_coords_count = ((df_result['경도'] == 0) & (df_result['위도'] == 0)).sum()
    
    print(f"📂 {region} 데이터 저장 완료: {output_path}")
    print(f"⚠️ {region}에서 위경도 변환 실패한 데이터 개수: {zero_coords_count}개\n")

print("🎉 모든 변환 및 저장 완료!")

```

```python
import pandas as pd
import requests
import json
import re

# 🔹 주소 → 경도, 위도 변환 함수
def doro_to_coords(df):
    data = df.copy().reset_index(drop=True)
    data['경도'] = 0
    data['위도'] = 0
    
    for i in range(len(data)):
        if pd.notna(data.iloc[i][7]):  # 도로명주소가 존재할 때만 변환 시도 (7번째 컬럼이 도로명주소)
            try:
                address = data.iloc[i][7]  # 7번째 컬럼(도로명주소)
                url = f'https://dapi.kakao.com/v2/local/search/address.json?query={address}'
                headers = {"Authorization": "KakaoAK " + "39aa05f286da07187ee24e01563aa0f2"}
                response = requests.get(url, headers=headers)
                response_json = json.loads(response.text.encode('utf-8').decode('utf-8'))

                if response_json.get('documents'):
                    temp = response_json['documents'][0].get('road_address') or response_json['documents'][0].get('address')
                    if temp:
                        data.loc[i, ['경도', '위도']] = [temp['x'], temp['y']]
            except Exception as e:
                print(f"❌ 오류 발생: {e} (행 {i})")
    
    return data

# 🔹 파일 읽기 (전체 행을 읽고 첫 8개 칼럼만 사용)
file_path = r"C:\Users\user\Desktop\0221 미팅\서울시 학교 기본정보.csv"
df = pd.read_csv(file_path, encoding='cp949', header=None)

# 🔹 처음 8개의 칼럼만 선택
df = df.iloc[:, :8]

# 🔹 도로명주소에서 자치구 추출하기 (강남구, 마포구, 송파구)
df['자치구'] = df[7].apply(lambda x: '강남구' if '강남구' in str(x) 
                                     else ('마포구' if '마포구' in str(x) 
                                           else ('송파구' if '송파구' in str(x) else None)))

# 🔹 자치구별로 필터링
df_filtered = df[df['자치구'].isin(['강남구', '마포구', '송파구'])].copy()

# 🔹 주소 → 경도, 위도 변환
df_result = doro_to_coords(df_filtered)

# 🔹 위경도 변환 실패한 데이터 개수 확인
zero_coords_count = ((df_result['경도'] == 0) & (df_result['위도'] == 0)).sum()

# 🔹 각 구별로 중복된 학교 제거
for region in ['강남구', '마포구', '송파구']:
    region_df = df_result[df_result['자치구'] == region]
    # 도로명주소(7번째 컬럼) 기준으로 중복 제거
    region_df = region_df.drop_duplicates(subset=[3])  
    region_df.to_csv(f"C:\\Users\\user\\Desktop\\0221 미팅\\{region}_학교정보.csv", encoding='utf-8-sig', index=False)
    print(f"📂 {region} 데이터 저장 완료")
    
print(f"⚠️ 위경도 변환 실패한 데이터 개수: {zero_coords_count}개")

```

```python
import pandas as pd
import requests
import json
import re

# 🔹 주소 → 경도, 위도 변환 함수
def doro_to_coords(df):
    data = df.copy().reset_index(drop=True)
    data['경도'] = 0
    data['위도'] = 0
    
    for i in range(len(data)):
        if pd.notna(data.iloc[i]['시설주소']):  # 시설주소가 존재할 때만 변환 시도
            try:
                address = data.iloc[i]['시설주소']
                url = f'https://dapi.kakao.com/v2/local/search/address.json?query={address}'
                headers = {"Authorization": "KakaoAK " + "39aa05f286da07187ee24e01563aa0f2"}
                response = requests.get(url, headers=headers)
                response_json = json.loads(response.text.encode('utf-8').decode('utf-8'))

                if response_json.get('documents'):
                    temp = response_json['documents'][0].get('road_address') or response_json['documents'][0].get('address')
                    if temp:
                        data.loc[i, ['경도', '위도']] = [temp['x'], temp['y']]
            except Exception as e:
                print(f"❌ 오류 발생: {e} (행 {i})")
    
    return data

# 🔹 파일 읽기
file_path = r"C:\Users\user\Desktop\0221 미팅\서울시 사회복지시설 목록.csv"
df = pd.read_csv(file_path, encoding='cp949')

# 🔹 시군구명 & 시설주소 칼럼 결측치 확인
missing_sgg = df['시군구명'].isna().sum()
missing_address = df['시설주소'].isna().sum()
print(f"⚠️ 시군구명 결측치 개수: {missing_sgg}개")
print(f"⚠️ 시설주소 결측치 개수: {missing_address}개")

# 🔹 시설주소가 결측치이고, 시군구명이 강남구/마포구/송파구인 데이터 확인
missing_address_filtered = df[df['시설주소'].isna() & df['시군구명'].isin(['강남구', '마포구', '송파구'])]

# 🔹 결과 출력
if not missing_address_filtered.empty:
    print("⚠️ 시설주소가 결측치이고, 시군구명이 강남구/마포구/송파구인 데이터가 있습니다!")
    print(missing_address_filtered)
else:
    print("✅ 시설주소가 결측치인 데이터 중 강남구, 마포구, 송파구에 해당하는 데이터는 없습니다.")

# 🔹 강남구, 마포구, 송파구 데이터 필터링
df_filtered = df[df['시군구명'].isin(['강남구', '마포구', '송파구'])].copy()

# 🔹 시설주소에서 괄호 제거
df_filtered['시설주소'] = df_filtered['시설주소'].apply(lambda x: re.sub(r'\s*\(.*?\)$', '', str(x)) if pd.notna(x) else x)

# 🔹 주소 → 경도, 위도 변환
df_result = doro_to_coords(df_filtered)

# 🔹 위경도 변환 실패한 데이터 개수 확인
zero_coords_count = ((df_result['경도'] == 0) & (df_result['위도'] == 0)).sum()

# 🔹 강남구, 마포구, 송파구별로 데이터 저장
for region in ['강남구', '마포구', '송파구']:
    region_df = df_result[df_result['시군구명'] == region]
    region_df.to_csv(f"C:\\Users\\user\\Desktop\\0221 미팅\\{region}_사회복지시설.csv", encoding='utf-8-sig', index=False)
    print(f"📂 {region} 데이터 저장 완료")

print(f"⚠️ 위경도 변환 실패한 데이터 개수: {zero_coords_count}개")
```

```python
# 🔹 위경도가 0인 데이터 개수 & 총 데이터 개수 & 비율 계산
summary = {}

for region in ['강남구', '마포구', '송파구']:
    total_count = df_result[df_result['시군구명'] == region].shape[0]  # 해당 구의 총 데이터 개수
    zero_coords_count = df_result[(df_result['시군구명'] == region) & (df_result['위도'] == 0) & (df_result['경도'] == 0)].shape[0]  # 위경도가 0인 개수
    zero_coords_ratio = (zero_coords_count / total_count) * 100 if total_count > 0 else 0  # 비율 (%)

    summary[region] = {
        "총 데이터 개수": total_count,
        "위경도 0 개수": zero_coords_count,
        "비율 (%)": round(zero_coords_ratio, 2)
    }

# 🔹 결과 출력
for region, stats in summary.items():
    print(f"📍 {region}")
    print(f"   ▶ 총 데이터 개수: {stats['총 데이터 개수']}개")
    print(f"   ▶ 위경도 0 개수: {stats['위경도 0 개수']}개")
    print(f"   ▶ 비율: {stats['비율 (%)']}%\n")

```

```python
import requests
import pandas as pd
import numpy as np

# 도로명주소를 위경도로 변환하는 함수
def doro_to_coords(df):
    data = df.copy()
    data['경도'] = 0
    data['위도'] = 0
    
    for i in range(len(data)):
        if pd.notna(data.loc[i, '도로명주소']) and data.loc[i, '도로명주소'] != '탈퇴':  # '도로명주소' → '주소' 변경
            try:
                address = data.loc[i, '도로명주소']
                url = f'https://dapi.kakao.com/v2/local/search/address.json?query={address}'
                headers = {"Authorization": "KakaoAK " + "39aa05f286da07187ee24e01563aa0f2"}  # 본인의 Kakao API 키
                response = requests.get(url, headers=headers).json()
                
                if response.get('documents'):
                    temp = response['documents'][0].get('road_address')  # 도로명주소 정보 가져오기
                    if temp:
                        data.loc[i, ['경도', '위도']] = [temp['x'], temp['y']]
            
            except Exception as e:
                print(f"오류 발생: {e} (행 {i})")
    
    return data

# CSV 파일 읽기 (경로 수정)
df = pd.read_csv(r"C:\Users\user\Desktop\0221 미팅\서울시 지하철 승하차 전처리.csv", encoding='utf-8-sig')

# 함수 적용
result_df = doro_to_coords(df)

# 각 구별로 데이터 나누기
gangnam = result_df[result_df['도로명주소'].str.contains("강남구", na=False)]
mapo = result_df[result_df['도로명주소'].str.contains("마포구", na=False)]
songpa = result_df[result_df['도로명주소'].str.contains("송파구", na=False)]

# 각 구별로 위경도가 0인 데이터 개수와 비율 계산
def calculate_missing_coords(data):
    missing_coords = data[(data['경도'] == 0) & (data['위도'] == 0)]
    missing_count = len(missing_coords)
    total_count = len(data)
    missing_ratio = missing_count / total_count * 100  # 비율 계산
    return missing_count, missing_ratio

# 강남구, 마포구, 송파구의 결측치 개수와 비율 확인
gangnam_missing_count, gangnam_missing_ratio = calculate_missing_coords(gangnam)
mapo_missing_count, mapo_missing_ratio = calculate_missing_coords(mapo)
songpa_missing_count, songpa_missing_ratio = calculate_missing_coords(songpa)

print(f"강남구 결측치 개수: {gangnam_missing_count}, 비율: {gangnam_missing_ratio:.2f}%")
print(f"마포구 결측치 개수: {mapo_missing_count}, 비율: {mapo_missing_ratio:.2f}%")
print(f"송파구 결측치 개수: {songpa_missing_count}, 비율: {songpa_missing_ratio:.2f}%")

# 각 구별로 파일 저장
gangnam.to_csv(r"C:\Users\user\Desktop\0221 미팅\강남구_서울시 지하철 승하차 위경도.csv", encoding='utf-8-sig', index=False)
mapo.to_csv(r"C:\Users\user\Desktop\0221 미팅\마포구_서울시 지하철 승하차 위경도.csv", encoding='utf-8-sig', index=False)
songpa.to_csv(r"C:\Users\user\Desktop\0221 미팅\송파구_서울시 지하철 승하차 위경도.csv", encoding='utf-8-sig', index=False)
```
