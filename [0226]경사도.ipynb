```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 1. 데이터 불러오기
gangnam_slp = gpd.read_file("/content/drive/MyDrive/25-1R/project/dataset/경사도(5179)/gangnam_slp_stat.shp")
songpa_slp = gpd.read_file("/content/drive/MyDrive/25-1R/project/dataset/경사도(5179)/songpa_slp_stat.shp")
mapo_slp = gpd.read_file("/content/drive/MyDrive/25-1R/project/dataset/경사도(5179)/mapo_slp_stat.shp")

# # 2. 데이터 크기 확인
# print("Gangnam Shape:", gangnam_slp.shape)
# print("Songpa Shape:", songpa_slp.shape)
# print("Mapo Shape:", mapo_slp.shape)

# 3. 시각화 (slp_mean 값 기준으로 색상 지정)
fig, ax = plt.subplots(1, 1, figsize=(12, 10))

# Gangnam 시각화
gangnam_slp.plot(
    column='slp_mean',
    cmap='viridis',
    legend=True,
    legend_kwds={
        'label': "Gangnam Slope Mean",
        'shrink': 0.5  # 범례 크기를 축소 (0.5는 기본 크기의 50%)
    },
    ax=ax
)

# Songpa 시각화
songpa_slp.plot(
    column='slp_mean',
    cmap='plasma',
    legend=True,
    legend_kwds={
        'label': "Songpa Slope Mean",
        'shrink': 0.5
    },
    ax=ax
)

# Mapo 시각화
mapo_slp.plot(
    column='slp_mean',
    cmap='coolwarm',
    legend=True,
    legend_kwds={
        'label': "Mapo Slope Mean",
        'shrink': 0.5
    },
    ax=ax
)

# 4. 제목 및 축 제거
ax.set_title("Slope Mean Visualization", fontsize=15)
ax.axis('off')  # 축 제거

# 5. 시각화 출력
plt.show()
```

## 경사 기반 후보지 필터링

```python
import geopandas as gpd

# 경사도 기준 적용
gangnam_filter = gangnam_slp[gangnam_slp['slp_mean'] <= 15]  # slp_mean <= 15인 격자만 유지
songpa_filter = songpa_slp[songpa_slp['slp_mean'] <= 15]  # slp_mean <= 15인 격자만 유지
mapo_fliter = mapo_slp[mapo_slp['slp_mean'] <= 15]  # slp_mean <= 15인 격자만 유지


# 시각화
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

mapo_fliter.plot(ax=ax[0], color='blue', edgecolor='black')
ax[0].set_title('Mapo District')

songpa_filter.plot(ax=ax[1], color='green', edgecolor='black')
ax[1].set_title('Songpa District')

gangnam_filter.plot(ax=ax[2], color='red', edgecolor='black')
ax[2].set_title('Gangnam District')

plt.tight_layout()
plt.show()
```
