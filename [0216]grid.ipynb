```python
import geopandas as gpd

# 1. 데이터 불러오기
mapo_file = '/content/drive/MyDrive/25-1R/project/dataset/(B100)국토통계_인구정보-총 인구 수(전체)-(격자) 100M_서울특별시 마포구_202410/nlsp_021001001.shp'
test_file = '/content/drive/MyDrive/25-1R/project/dataset/11440/11440.shp'

mapo = gpd.read_file(mapo_file, columns='geometry')
test = gpd.read_file(test_file, columns='geometry')

# 좌표계 확인 및 통일
if mapo.crs is None:
    mapo.set_crs(epsg=5181, inplace=True)  # EPSG:5181 (GRS80 UTM-K)
if test.crs != mapo.crs:
    test = test.to_crs(mapo.crs)

# 2. 교차 분석 (mapo 격자와 test 데이터 매핑)
intersected_mapo = gpd.overlay(mapo, test, how='intersection')

# 3. 교차 영역 면적 계산
intersected_mapo['intersection_area'] = intersected_mapo.geometry.area
intersected_mapo.head()
```

포함 여부 열 추가

```python
import geopandas as gpd

# 1. 데이터 불러오기
gangnam_file = '/content/drive/MyDrive/25-1R/project/dataset/(B100)국토통계_인구정보-총 인구 수(전체)-(격자) 100M_서울특별시 강남구_202410/nlsp_021001001.shp'
mt_gangnam_file = '/content/drive/MyDrive/25-1R/project/dataset/11680/11680.shp'

gangnam = gpd.read_file(gangnam_file, columns='geometry')
mt_gangnam = gpd.read_file(mt_gangnam_file, columns='geometry')

# 좌표계 확인 및 통일
if gangnam.crs is None:
    gangnam.set_crs(epsg=5181, inplace=True)  # EPSG:5181 (GRS80 UTM-K)
if mt_gangnam.crs != gangnam.crs:
    mt_gangnam = mt_gangnam.to_crs(gangnam.crs)

# 2. 공간 조인 (gangnam과 mt_gangnam 데이터 교차 여부 확인)
# how='left': gangnam 기준으로 조인
# op='intersects': mt_gangnam과 교차하는지 확인
joined = gpd.sjoin(gangnam, mt_gangnam, how='left', predicate='intersects').drop_duplicates()

joined.head()
# # 3. 'mt' 열 추가 (교차 여부에 따라 값 설정)
# # 교차된 경우: 1, 교차되지 않은 경우: 0
# gangnam['mt'] = joined['index_right'].notnull().astype(int)

# # 결과 확인
# gangnam[['geometry', 'mt']].head()
```

```python
gangnam.shape
```

```python
joined.shape
```

how='left' 옵션을 사용했으므로, gangnam을 기준으로 모든 데이터를 포함하며, 교차하는 경우에는 mt_gangnam의 정보를 추가합니다.

하지만, 한 gangnam 지오메트리가 여러 mt_gangnam 지오메트리와 교차하면 해당 gangnam 행이 결과 데이터프레임에 여러 번 나타납니다.

```python
joined = gpd.sjoin(gangnam, mt_gangnam, how='left', predicate='intersects')
joined = joined[~joined.index.duplicated(keep='first')]

# 3. 'mt' 열 추가 (교차 여부에 따라 값 설정)
# 교차된 경우: 1, 교차되지 않은 경우: 0
gangnam['mt'] = joined['index_right'].isnull().astype(int)

# 결과 확인
gangnam[['geometry', 'mt']].head()
```

```python
import matplotlib.pyplot as plt

# 시각화 크기 설정
plt.figure(figsize=(10, 10))

# gangnam을 시각화 (mt가 1인 경우는 색칠, 0인 경우는 투명)
ax = gangnam.plot(
    color='white', edgecolor='black', figsize=(10, 10), alpha=0.5, linewidth=0.5
)
# mt가 1인 경우 내부를 빨간색으로 색칠
gangnam[gangnam['mt'] == 1].plot(
    ax=ax, color='red', alpha=0.6, edgecolor='black', linewidth=0.5
)

# 제목 및 축 제거
plt.title("Visualization of Gangnam with 'mt' = 1", fontsize=16)
plt.axis('off')

# 그래프 출력
plt.show()
```
