```python
pm_acc = pm[['발생년월', '시군구', '사망자수', '중상자수', '경상자수', '부상신고자수']]
pm_acc.head()
```

```python
pm_acc.info()
```

```python
import pandas as pd


# 1. '사고건수' 열 생성 ('사망자수', '중상자수', '경상자수', '부상신고자수'의 합)
pm_acc['사고건수'] = pm_acc[['사망자수', '중상자수', '경상자수', '부상신고자수']].sum(axis=1)

# 2. '시군구'에서 중간의 '구' 정보를 추출하여 '자치구' 열 생성
pm_acc['자치구'] = pm_acc['시군구'].str.extract(r'([가-힣]+구)').squeeze()

pm_acc
```

```python
pm_acc.loc[:,'자치구'].value_counts()
```

```python
# 자치구별 사고건수를 집계
pm_acc_sort = pm_acc.groupby('자치구')['사고건수'].sum().reset_index()

pm_acc_sort = pm_acc_sort.sort_values(by='사고건수', ascending=False).reset_index(drop=True)
pm_acc_sort
```

```python
import pandas as pd

# 데이터 정의
pm_trac = {
    '자치구': ['마포구', '송파구', '성동구', '강남구', '영등포구', '서초구', '동작구', '관악구',
             '용산구', '동대문구', '강서구', '광진구', '중랑구', '강동구', '노원구', '도봉구',
             '서대문구', '중구', '종로구', '강북구', '성북구', '구로구', '양천구', '은평구', '금천구'],
    '견인건수': [12588, 12234, 11163, 10802, 10257, 8965, 7128, 6632,
          5766, 5408, 5392, 4174, 3846, 3640, 3404, 3287,
          2863, 2663, 2148, 2098, 2073, 1628, 1376, 387, 10]
}

# 데이터프레임 생성
pm_trac = pd.DataFrame(pm_trac)

pm_trac
```

```python
# 상위 5개 자치구 추출
top_5_pm_acc = pm_acc_sort.head(5)['자치구']
top_5_pm_trac = pm_trac.sort_values(by='견인건수', ascending=False).head(5)['자치구']

# 공통 자치구 확인
common_gu = set(top_5_pm_acc).intersection(set(top_5_pm_trac))
print("공통 자치구:", common_gu)
```
