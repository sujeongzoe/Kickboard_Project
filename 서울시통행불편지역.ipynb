```python
import requests
import json
import pandas as pd

# 빈 리스트
all_data = []

# 1000개씩 데이터 불러오기
for start in range(1, 14000, 1000):
    url = f'http://openapi.seoul.go.kr:8088/45596e627172616435346f7a486f4e/json/passWeak/{start}/{start+999}/'
    content = requests.get(url).json()
    all_data.extend(content.get('passWeak', {}).get('row', []))

# 하나의 데이터프레임으로 변환
result = pd.DataFrame(all_data)
result
```

```python
import re
import pandas as pd

# 정규표현식으로 주소만 추출하는 함수
def extract_address(address):
    # '한글 + 숫자-숫자' 형식만 추출 (우, 좌, 번지 아래 등은 제외)
    match = re.match(r'([가-힣]+\s?\d{1,5}-\d{1,5})', address)
    if match:
        return match.group(1)  # 매칭된 주소 반환
    return address  # 매칭되지 않으면 원본 반환

# '통행불편지역이름' 컬럼에 함수 적용하여 새로운 주소 컬럼 생성
result['주소'] = result['통행불편지역이름'].apply(extract_address)
# 결과 출력
print(result)

# 엑셀 파일로 저장
file_path = r'C:\Users\user\Desktop\dataset\서울시통행불편지역_주소추출.xlsx'
result.to_excel(file_path, index=False)
```
